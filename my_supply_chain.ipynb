```python
import random
import matplotlib.pyplot as plt

A = 10000
B = 10000
C = 0
D = 0

p = 0.9
time_steps = 15000

A_values = []
B_values = []
C_values = []
D_values = []
T_values = []

for t in range(time_steps):
    if A > 0 and B > 0:
        if random.random() < p:
            A -= 1
            B -= 1
            C += 2
            D += 1
    
    if t % 100 == 0:
        A_values.append(A)
        B_values.append(B)
        C_values.append(C)
        D_values.append(D)
        T_values.append(t)

plt.scatter(T_values, A_values, label='A')
plt.scatter(T_values, B_values, label='B')
plt.scatter(T_values, C_values, label='C')
plt.scatter(T_values, D_values, label='D')
plt.legend()
plt.show()
```
